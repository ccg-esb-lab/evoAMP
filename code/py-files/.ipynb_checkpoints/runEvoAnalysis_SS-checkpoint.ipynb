```python

params=dict()
params['fileDataNames']=[('{0}/{1}').format(dirDataName,i) for i in fileDataNames_SS]
params['fileLayoutName']='%s/%s'%(dirDataName, fileLayoutName)
params['fileDictName']='%s/%s'%(dirDataName, fileDictName)
params['fileInoculationName']='%s/%s'%(dirDataName, fileInoculationName)
params['ignoreWells']=ignoreWells_SS
params['conv2ml']=conv2ml
params['verbose']=verbose

#IMPORTS TREATMENT DICT
treatment_dict=loadTreatmentDict(params) #READ DICTIONARY
  
#IMPORTS PLATE LAYOUT
M=importPlateLayout(params['fileLayoutName'], params) #READ RANDOMIZED PLATE LAYOUT
if params['verbose']:
    print('\n\n**** LAYOUT:\n')
    show(M)
    
#IMPORTS INOCULATION LAYOUT
B=importInoculationLayout(params)
if params['verbose']:
    print('\n\n**** Inoculation:\n')
    show(B)
```

```python

#LOADS SS-DATA
DATA_SS=[]
for n, fileDataNames in enumerate(params['fileDataNames']):
    OD=importData([fileDataNames])
        
    ignore_wells=params['ignoreWells'][n].split(',')
    OD=ignoreWells(OD, ignore_wells)
    
    thisDATA=compileData(M, OD, B)
    bg=getMeanODs(thisDATA, 'M','0')[0]
    thisDATA=subtractBG(thisDATA, bg) 
    
    if params['verbose']:
        print('\n\n**** %s:\n'%fileDataNames)
        show(OD)
        print('bg:%s'%bg)
    
    DATA_SS.append(thisDATA)

```
