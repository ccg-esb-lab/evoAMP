## Load Data

```python
#IMPORTS TREATMENT DICT
treatment_dict=loadTreatmentDict(params) #READ DICTIONARY
  
#IMPORTS PLATE LAYOUT
M=importPlateLayout(params['fileLayoutName'], params) #READ RANDOMIZED PLATE LAYOUT
if params['verbose']:
    print('\n\n**** LAYOUT:\n')
    show(M)
    
#IMPORTS INOCULATION LAYOUT
B=importInoculationLayout(params)
if params['verbose']:
    print('\n\n**** Inoculation:\n')
    show(B)
```

```python

#LOADS LS-DATA
DATA_LS=[]
for fileDataNames in params['fileDataNames_LS']:
    OD=importData([fileDataNames])
    thisDATA=compileData(M, OD, B)
    bg=getMeanODs(thisDATA, 'M','0')[0]
    thisDATA=subtractBG(thisDATA, bg) 
    
    if params['verbose']:
        print('\n\n**** %s:\n'%fileDataNames)
        show(OD)
        print('bg:%s'%bg)
    
    DATA_LS.append(thisDATA)


#LOADS SS-DATA
DATA_SS=[]
for fileDataNames in params['fileDataNames_SS']:
    OD=importData([fileDataNames])
    thisDATA=compileData(M, OD, B)
    bg=getMeanODs(thisDATA, 'M','0')[0]
    thisDATA=subtractBG(thisDATA, bg) 
    
    if params['verbose']:
        print('\n\n**** %s:\n'%fileDataNames)
        show(OD)
        print('bg:%s'%bg)
    
    DATA_SS.append(thisDATA)
```
